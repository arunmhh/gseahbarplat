```r
library(tidyverse)
library(ggplot2)
library(ggthemes)
library(ggpubr)
```

```r
zdhhc17 <- read.csv("/Volumes/NGS/LCSC/.ipynb_checkpoints/epcm_code/Control_ZDHHC3.csv", header = TRUE, sep = ",")
head(zdhhc17)
```

```r
df_long <- zdhhc17 %>% 
  pivot_longer(
    cols = 2:22,  # Columns to pivot
    names_to = "gene",  # Name for new column that will contain score variable names
    values_to = "log2_SI"  # Name for new column that will contain score values
  )
  df_long
```

```r
df_long %>% group_by(sample, gene)   -> pvl 
```

```r
summarize(pvl, mean_fc = mean(log2_SI), sd_fc = sd(log2_SI)) -> df1
df1 
```

```r
for (gene in unique(df1$gene)) {
  df_gene <- subset(df1, gene == gene)
  p <- ggplot(df_gene, aes(x = sample, y = mean_fc)) +
    geom_point(size = 3, position = position_dodge (1)) +
    geom_errorbar(aes(ymin = mean_fc - sd_fc, ymax = mean_fc + sd_fc), width = 0.2) +
    ggtitle(paste("Gene:", gene)) +
    xlab("Sample Type") +
    ylab("Fold Change") +
    theme_bw()
  ggsave(filename = paste0("/Volumes/NGS/projectVSC/snakemakelearning/visuals/zdhhc/",gene, "_dotplot.png"), plot = p, dpi = 300)
}
```

```r
zz  <- ggplot(df_gene, aes(x = sample, y = log2_SI, color = gene))+
    geom_point(size = 3, position = position_dodge (1))+
    facet_wrap(~gene, ncol = 7)+
    ggtitle("ZDHHC3")+
    xlab("Sample Type")+
    ylab("Fold Change")+
    theme_light()+
    theme(axis.text.x = element_text(angle = 90, hjust = 1, vjust = 0.5))+
    theme(legend.position = "none")+
    theme(axis.ticks = element_blank(),plot.title = element_text(margin = margin(b=10),color = "red", size = 20,face = "bold"))
    zz
```

```r
library(ggplot2)

# read in your data
#mydata <- read.csv("mydata.csv")

# loop through each unique gene in the data
for (gene in unique(df_long$gene)) {
  
  # create a data frame for just this gene
  this_gene_data <- df_long[df_long$gene == gene,]
  
  # create a ggplot dot plot for this gene
  p <- ggplot(this_gene_data, aes(x = sample, y = log2_SI)) +
         geom_dotplot(binaxis = "y",stackdir = "center", fill = "black") +
         labs(title = gene) +
         theme_classic()+
         ylim (0 , 10)+
         stat_summary(fun.y = mean, geom = "point",shape = 23, size = 3, fill = "white", color = "black")+
         stat_compare_means(method = "t.test", comparisons = list(c("ASO_ZDHHC17", "Scramble cont.")), label = "p.format", hide.ns = TRUE)+
         theme(axis.line = element_line(linetype = "solid"),
    panel.grid.major = element_line(linetype = "blank"),
    panel.grid.minor = element_line(linetype = "blank"),
    axis.title = element_text(face = "bold"),
    axis.text.x = element_text(family = "mono",size = 15, face = "bold"), 
    axis.text.y = element_text(family = "mono",size = 15, face = "bold"),
    plot.title = element_text(size = 15, colour = "brown1", hjust = 0.5),
    legend.title = element_text(family = "mono")) +
    labs(title = (paste0( gene, " Kinase Response in Control and ZDHHC3 ")),
                x = (paste0(gene," Kinase")),
                y = "Log2 Signal Intensity", colour = "Black",
                fill = NULL) + theme(axis.text = element_text(size = 7))
  # save the plot as a PNG file with a filename based on the gene name
  ggsave(paste0("/Volumes/NGS/projectVSC/snakemakelearning/visuals/dotplot/", gene, ".png"), plot = p, width = 6, height = 4)
}
```
